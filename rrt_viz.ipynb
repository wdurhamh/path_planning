```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_circle(ax, c, r=1):
    thetas = np.arange(0, 2*np.pi+0.1, 0.1)
    plt.plot(c[0] + r*np.cos(thetas), c[1]+r*np.sin(thetas))
```

```python
with open("tree.txt","r") as f:
    for line in f:
        vals = line.strip().split(",")
        plt.plot([float(vals[0]), float(vals[2])], [float(vals[1]), float(vals[3])])
plot_circle(plt.gca(), [3,2.5])
plot_circle(plt.gca(), [4,8])
plot_circle(plt.gca(), [7,5])
plot_circle(plt.gca(), [9,5])
plt.gca().set_aspect("equal")
```
